## Using an MLP model to predict housing prices based on basic features and distance
### Some Notes: MAE before addition of location is ~ 218768
### With distance to CBD and suburb ID its down to ~ 180000

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/base_info/current_with_distance.csv")
#print(df[["Price", "Beds", "Bathrooms", "Carports", "Size", "CBD_Dist", "Suburb"]])
labels, levels = pd.factorize(df['Suburb'])
df["Suburb_ID"] = labels
print(df)
```

```python
size_cut = df[df['Size'] != -1]
car_cut = size_cut[size_cut['Carports'] != -1]
bath_cut = car_cut [car_cut ['Bathrooms'] != -1]
dist_cut = bath_cut[bath_cut['CBD_Dist'] != -1]
final_cut = dist_cut[["Link", "Price", "Beds", "Bathrooms", "Carports", "Size", "CBD_Dist","Suburb_ID"]]
answers = final_cut["Price"].values.tolist()
data = final_cut[["Beds", "Bathrooms", "Carports", "Size", "CBD_Dist","Suburb_ID"]].values.tolist()
print(data)
print(answers)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

X_train, X_test, y_train, y_test = train_test_split(data, answers, random_state=1)
regr = MLPRegressor(random_state=2).fit(X_train, y_train)
pred = list(regr.predict(X_test))
mean_absolute_error(np.array(pred), np.array(y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression

regr = RandomForestRegressor(max_depth=20, n_estimators=1000, random_state=1)
regr.fit(X_train, y_train)
pred = list(regr.predict(X_test))
mean_absolute_error(np.array(pred), np.array(y_test))
```

```python
print(pred,y_test)
```

```python
mae = []
actuals = []
for row in final_cut.values.tolist():
    print(row)
    price = row[1]
    inputdata = row[2:]
    pred = regr.predict([inputdata])
    print("The real price was " + str(price) + " and the model predicted " + str(pred))
    mae.append((pred-price))
    actuals.append(price)
```

```python
#print(final_cut["Link"].values)
#print(len(final_cut["Link"].values))
#print([x[0] for x in mae])
#print(len([x[0] for x in mae]))
data = list(zip(final_cut["Link"].values, [x[0] for x in mae], actuals))
sorted_by_second = sorted(data, key=lambda tup: tup[1], reverse = True)
print(sorted_by_second)
```
